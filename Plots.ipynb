```python
import numpy as np
import pandas as pd
# we want our plots inline
%matplotlib inline
# used for seeding random number sequences
seedval = 111111
```


# Histograms
Histograms are useful for visualizing distributions of data. The following shows you a histogram of generating 1000 values from the normal distribution:

```python
np = [-5.5,-5,-4.15,-4.85,-6.05,-5,-4.1,-4.5,-4.6]
df = pd.DataFrame(np)
df.hist(bins=5)
```

```python
# Create a histogram
np.random.seed(seedval)
# 1000 random numbers
df = pd.DataFrame(np.random.randn(1000))
# Generate the histogram plot
df.hist()
```

Histogram again, but with more bins

```python
df.hist(bins = 100);
```

Which histogram represents the data better? Why?

# Box Plot

```python
# Generate the series
np.random.seed(seedval)
df = pd.DataFrame(np.random.randn(10,5))
# Generate the box plot
df.boxplot(return_type='axes');
```

Are there any outliers?

# Bar Chart

```python
# Generate the series
# create a small series of 10 random values 
ds = pd.Series(np.random.rand(10))
# Generate the bar chart
ds.plot(kind='bar');
```

# Scatter Plot

```python
# Generate two series of normally distributed random values
# We would expect this to cluster around 0,0
df = pd.DataFrame(np.random.randn(10, 2), 
                        columns=['a', 'b'])
# Generate the scatter plot
df.plot(kind='scatter', x='a', y='b')
```

```python
df.plot()
```

```python
df.plot(kind='bar')
```
